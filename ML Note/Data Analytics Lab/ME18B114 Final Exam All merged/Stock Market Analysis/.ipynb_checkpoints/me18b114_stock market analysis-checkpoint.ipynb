## Formatting the Date format

```python
#cognizant

cognizant['Date'] = pd.to_datetime(cognizant['Date'],format='%Y-%m-%d')
cognizant['year'] = pd.DatetimeIndex(cognizant['Date']).year
cognizant['month'] = pd.DatetimeIndex(cognizant['Date']).month
```

```python
#hcl

hcl['Date'] = pd.to_datetime(hcl['Date'],format='%Y-%m-%d')
hcl['year'] = pd.DatetimeIndex(hcl['Date']).year
hcl['month'] = pd.DatetimeIndex(hcl['Date']).month
```

```python
#hdfc

hdfc['Date'] = pd.to_datetime(hdfc['Date'],format='%Y-%m-%d')
hdfc['year'] = pd.DatetimeIndex(hdfc['Date']).year
hdfc['month'] = pd.DatetimeIndex(hdfc['Date']).month
```

```python
#sbi

sbi['Date'] = pd.to_datetime(sbi['Date'],format='%Y-%m-%d')
sbi['year'] = pd.DatetimeIndex(sbi['Date']).year
sbi['month'] = pd.DatetimeIndex(sbi['Date']).month

```

```python
#icici

icici['Date'] = pd.to_datetime(icici['Date'],format='%Y-%m-%d')
icici['year'] = pd.DatetimeIndex(icici['Date']).year
icici['month'] = pd.DatetimeIndex(icici['Date']).month
```

```python
#infosys

infosys['Date'] = pd.to_datetime(infosys['Date'],format='%Y-%m-%d')
infosys['year'] = pd.DatetimeIndex(infosys['Date']).year
infosys['month'] = pd.DatetimeIndex(infosys['Date']).month
```

```python
#exchange

exchange_rate['Date'] = pd.to_datetime(exchange_rate['Date'],format='%Y-%m-%d')
exchange_rate['year'] = pd.DatetimeIndex(exchange_rate['Date']).year
exchange_rate['month'] = pd.DatetimeIndex(exchange_rate['Date']).month
```

## Returns


```python
sbi['Returns'] = (sbi['Close']/sbi['Close'].shift(1))-1
hcl['Returns'] = (hcl['Close']/hcl['Close'].shift(1))-1
hdfc['Returns'] = (hdfc['Close']/hdfc['Close'].shift(1))-1
infosys['Returns'] = (infosys['Close']/infosys['Close'].shift(1))-1
icici['Returns'] = (icici['Close']/icici['Close'].shift(1))-1
cognizant['Returns'] = (cognizant['Close']/cognizant['Close'].shift(1))-1
```

```python
sbi.head()
```

```python
plt.figure(figsize=(14,10))
plt.title('Volume Traded Trend')
sns.lineplot(x='Date',y='Returns',data= cognizant, label='Cognizant')
sns.lineplot(x='Date',y='Returns',data= infosys, label='Infosys')
sns.lineplot(x='Date',y='Returns',data= sbi, label='SBI')
sns.lineplot(x='Date',y='Returns',data= hdfc, label='HDFC')
sns.lineplot(x='Date',y='Returns',data= icici, label= 'ICICI')
sns.lineplot(x='Date',y='Returns',data= hcl, label='HCL')
```

## Cumulative Returns

```python
sbi['Cumulative Return']=(1+sbi['Returns']).cumprod()
hdfc['Cumulative Return']=(1+hdfc['Returns']).cumprod()
hcl['Cumulative Return']=(1+hcl['Returns']).cumprod()
icici['Cumulative Return']=(1+icici['Returns']).cumprod()
infosys['Cumulative Return']=(1+infosys['Returns']).cumprod()
cognizant['Cumulative Return']=(1+cognizant['Returns']).cumprod()
```

```python
sbi
```

```python
plt.figure(figsize=(14,10))
plt.title('Cumulative Returns IT Trend')
sns.lineplot(x='Date',y='Cumulative Return',data= cognizant, label='Cognizant')
sns.lineplot(x='Date',y='Cumulative Return',data= infosys, label='Infosys')
sns.lineplot(x='Date',y='Cumulative Return',data= hcl, label='HCL')
```

```python
plt.figure(figsize=(14,10))
plt.title('Cumulative Returns Banks Trend')
sns.lineplot(x='Date',y='Cumulative Return',data= sbi, label='SBI')
sns.lineplot(x='Date',y='Cumulative Return',data= hdfc, label='HDFC')
sns.lineplot(x='Date',y='Cumulative Return',data= icici, label= 'ICICI')
```
